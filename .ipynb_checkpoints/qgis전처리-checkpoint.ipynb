#####  법정동, 행정동 매핑

```python
dongs = pd.read_csv("C:/Users/bhg/Desktop/2022빅콘/조사 데이터/법정동행정동매핑.csv",encoding='cp949')
dongs.rename(columns={'읍면동명':'읍면동_매핑'},inplace=True)
dongs
```

```python
mapping_dongs_cars= pd.merge(dongs, cars,how='outer',left_on='동리명',right_on='동리명_차량')
# mapping_dong[['읍면동_매핑','동리명','동리명_차량']]
mapping_dongs_cars.dropna(subset=['동리명_차량'],inplace=True)
mapping_dongs_cars.drop(['시군구명','동리명','Unnamed: 3'],axis=1,inplace=True)
mapping_dongs_cars

mapping_total = pd.merge(mapping_dongs_cars,pop,left_on='읍면동_매핑',right_on='읍면동_인구')
mapping_total.drop(['읍면동_매핑'],axis=1,inplace=True)
mapping_total
# 격자는 읍면동 기준!
# 결과 살펴보니 인구의 읍면동이 차량의 동리명보다 크다 -> 차량이 합쳐져서 각 동에 들어가야 함
# 그러나 이상하게 2개의 동리명이 3개의 읍면동으로 들어가는 경우 있어
# ex) 중동과 동백동(동리명)이 동백1,2,3동
# 그냥 각 읍면동 내 동리명 차량의 평균을 차량대수로 보는 것이 맞아보임
```

```python
mapping_total.to_csv('mapping_total.csv',encoding='cp949')
```

```python
mapping_total.columns
```

```python
sep_dong = ['동백1동','동백2동','동백3동','풍덕천1동','풍덕천2동','상현1동','상현2동','상현3동','죽전1동','죽전2동','죽전3동','영덕1동','영덕2동']
cal_cars = mapping_total.loc[~mapping_total['읍면동_인구'].isin(sep_dong)]
# mapping_total.groupby('읍면동_인구')['합계_차량', '승  용', '승  합', '화  물', '특  수', '관  용', '자가용', '영업용',].sum()
cal_cars = cal_cars.groupby('읍면동_인구')['합계_차량', '승  용', '승  합', '화  물', '특  수', '관  용', '자가용', '영업용',].sum()
cal_cars.reset_index(inplace=True)
```

```python
mapping_total.columns
```

```python
cal_cars = pd.merge(cal_cars,mapping_total[['읍면동_인구','내국인 계', '내국인 남', '내국인 여', '외국인 계', '외국인 남', '외국인 여', '합계_인구',
       ' 10대 미만 ', ' 10대 ', ' 20대 ', ' 30대 ', ' 40대 ', ' 50대 ', ' 60대 ',
       ' 70대 ', ' 80대 ', ' 90대 ', ' 90세 ', ' 100세 이상 ']],how='inner',on='읍면동_인구')
cal_cars
```
